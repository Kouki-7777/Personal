```python
def funcd(x):
    x: float
    f: float
    fd: float
    
    f = x**2 - 2.0
    fd = 2.0 * x
    return f, fd
```

```python
def newton(xmin, xmax, acc):
    xmin: float
    xmax: float
    acc: float
    root: float
    eflag: int
    IMAX: int
    
    #初期値の設定
    root = (xmin + xmax) / 2.0
    eflag = 0
    IMAX = 100
    
    for i in range(IMAX):
        #funcdからf(x)とf'(x) を取得
        f, fd = funcd(root)
        
        #古い値を保存
        root_old = root
        
        #ゼロ割り防止
        #微分係数が0に近い場合のエラーコード
        if abs(fd) < 1e-15:
            eflag = 3 
            return root, eflag
            
        # ニュートン法の更新式: x_{j} = x_{j-1} - f(x_{j-1}) / f'(x_{j-1})
        root = root_old - f / fd
        
        #収束条件：|x_j - x_{j-1}|<accで成功した場合
        if abs(root - root_old) < acc:
            eflag = 0
            return root, eflag
        
        #[xmin, xmax]の範囲を出てしまった場合
        if root < xmin or root > xmax:
            eflag = 1
            return root, eflag
            
        
            
    #IMAX回繰り返してもaccの精度に収束しなかった場合
    eflag = 2
    return root, eflag
```

```python
acc: float = 1.0e-6
x1: float = 1.0
x2: float = 2.0
root: float
eflag: int
root, eflag = newton(x1, x2, acc)
if eflag == 0:
    print("Root=", root)
else:
    print("Error=", eflag)
```
